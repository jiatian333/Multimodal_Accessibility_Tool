```python
import psycopg2
from psycopg2.extensions import AsIs
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt

```

```python

geojson_file = 'isochrones.geojson'

gdf = gpd.read_file(geojson_file)

print(gdf.head())
```

```python
print(gdf.columns)
level_column = gdf['level']
geometry_column = gdf['geometry']
print(level_column.head())
print(type(geometry_column[0]))
```

```python
fig, ax = plt.subplots(figsize=(10, 8))
gdf.plot(ax=ax, edgecolor="black", alpha=0.5, cmap="coolwarm")

ax.set_title("Fußgänger-Isokronen")
ax.set_xlabel("Längengrad")
ax.set_ylabel("Breitengrad")

plt.show()
```
